### Şimdi "Life Expectancy" ile "Unemployment Rate" arasındaki ilişkiyi hem görsel hem de makine öğrenmesi yöntemleriyle inceleyelim

```python
sns.set(style="whitegrid")
sns.lmplot(df,x="Unemployment Rate (%)",y="Life Expectancy",height=6,aspect=1.5)
plt.title("The Relationship Between Life Expectancy and Unemployment Rate")
plt.xlabel("İşssizlik Oranı %")
plt.ylabel("Beklenen Yaşam Süresi")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
X = df[["Unemployment Rate (%)"]]
y = df[["Life Expectancy"]]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
print(f"Katsayı (Coefficient): {model.coef_[0]}")
print(f"Sabit (Intercept): {model.intercept_}")
```

```python
y_pred = model.predict(X_train)
y_pred
```

```python
mse = mean_squared_error(y_train,y_pred)
r2 = r2_score(y_train,y_pred)
```

Life Expectancy ve Unemployment Rate arasında dağılım lineer olmadığından lineer regresyon kullanarak bir ilişki bulamadık SVR modeli deneyelim

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
```

```python
svr_model = SVR(kernel="rbf",C= 1.0,epsilon = 0.1)
```

```python
scaler_X = StandardScaler()
scaler_y = StandardScaler()
```

```python
X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)
y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1,1)).flatten()
```

```python
svr_model.fit(X_train_scaled,y_train_scaled)
```

```python
y_pred_scaled = svr_model.predict(X_test_scaled)
y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1,1)).flatten()
```

```python
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print("Mean Squared Error (MSE):", mse)
print("R^2 Score:", r2)
```

```python
import matplotlib.pyplot as plt
y_pred_scaled = svr_model.predict(scaler_X.transform(X))
y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1))
plt.plot(X,y_pred,color="red",label="SVR Prediction")
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.legend()
plt.show()
```
